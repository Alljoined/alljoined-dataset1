```python
from scipy.io import loadmat

def load_mat_file(file_path):
    mat_contents = loadmat(file_path)
    return mat_contents
```

```python
nsd_mat_file_path = 'nsd_expdesign.mat'
nsd_mat_contents = load_mat_file(nsd_mat_file_path)
```
